```python
# Loading the Average High Temperatures into a DataFrame
import pandas as pd

nyc = pd.read_csv('ave_hi_nyc_jan_1895-2018.csv')

# Renaming columns
nyc.columns = ['Date', 'Temperature', 'Anomaly']

# If 'Date' is stored as a string, convert it to an integer
nyc['Date'] = pd.to_numeric(nyc['Date'], errors='coerce')  # Converts to numbers, setting invalid parsing as NaN

# Apply floordiv if 'Date' is now numeric
nyc['Date'] = nyc['Date'].floordiv(100)

# Displaying the head and tail of the DataFrame
print("Head and Tail Output of the DataFrame showing January Temp")
print(nyc.head(5))
print()
print(nyc.tail(5))
```

```python
import matplotlib.pyplot as plt

#Convert Fahrenheit to Celsius (assuming Temperature is in Fahrenheit)
nyc['Celsius'] = (nyc['Temperature'] - 32) * 5/9

# Round both Fahrenheit and Celsius values to whole numbers
nyc['Temperature'] = nyc['Temperature'].round(0)
nyc['Celsius'] = nyc['Celsius'].round(0)

# Create a DataFrame with both Celsius and Fahrenheit values
df = pd.DataFrame({
    'Date': nyc['Date'],
    'Fahrenheit': nyc['Temperature'],
    'Celsius': nyc['Celsius']
})

# Show the first few rows of the DataFrame
print(df.head())

# Plot Celsius vs Fahrenheit
df.plot(x='Celsius', y='Fahrenheit', title="Celsius vs Fahrenheit", legend=True)

# Show the plot
plt.xlabel("Celsius")
plt.ylabel("Fahrenheit")
plt.grid(True)
plt.show()
```
